```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

dia_semana = [1, 2, 3, 4, 5, 6, 7]
num_clientes = [15, 20, 30, 100, 350, 500, 700]

dados_dias = {
    'Dia da Semana': dia_semana,
    'Nº de Clientes': num_clientes
}

df_dias = pd.DataFrame(dados_dias)
print("DataFrame:")
print(df_dias)
```

```python
matriz_dados_dias = df_dias[['Nº de Clientes']].to_numpy()

kmeans = KMeans(n_clusters=3, random_state=42)
df_dias['Cluster'] = kmeans.fit_predict(matriz_dados_dias)

print("\nDataFrame com Clusters:")
print(df_dias)

plt.scatter(df_dias['Dia da Semana'], df_dias['Nº de Clientes'], c=df_dias['Cluster'], cmap='viridis')
plt.title('Clusters de Dias de Maior Movimento')
plt.xlabel('Dia da Semana')
plt.ylabel('Nº de Clientes')
plt.show()
```
