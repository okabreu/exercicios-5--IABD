```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

setores = [1, 2, 3, 4, 5, 6, 7, 8]
produtos_fabricados = [100, 50, 15, 200, 500, 1000, 375, 450]

dados_setores = {
    'Setores': setores,
    'Nº de Produtos Fabricados': produtos_fabricados
}

df_setores = pd.DataFrame(dados_setores)
print("DataFrame:")
print(df_setores)
```

```python
matriz_dados_setores = df_setores[['Nº de Produtos Fabricados']].to_numpy()

kmeans = KMeans(n_clusters=3, random_state=42)
df_setores['Cluster'] = kmeans.fit_predict(matriz_dados_setores)

print("\nDataFrame com Clusters:")
print(df_setores)

plt.scatter(df_setores['Setores'], df_setores['Nº de Produtos Fabricados'], c=df_setores['Cluster'], cmap='viridis')
plt.title('Clusters de Setores Mais Produtivos')
plt.xlabel('Setores')
plt.ylabel('Nº de Produtos Fabricados')
plt.show()
```
