```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

dados = {
    'Substância': ['Álcool', 'Gasolina', 'Leite', 'Querosene', 'Óleo', 'Vinho'],
    'Concentração (%)': [12.5, 0.1, 4.0, 1.2, 0.5, 15.0],
    'Teor Alcoólico (%)': [50, 0.05, 0.01, 0.02, 0.01, 12.5]
}

df_substancias = pd.DataFrame(dados)
print("DataFrame:")
print(df_substancias)
```

```python
matriz_dados_substancias = df_substancias[['Concentração (%)', 'Teor Alcoólico (%)']].to_numpy()

kmeans = KMeans(n_clusters=2, random_state=42)
df_substancias['Cluster'] = kmeans.fit_predict(matriz_dados_substancias)

print("\nDataFrame com Clusters:")
print(df_substancias)

plt.scatter(df_substancias['Concentração (%)'], df_substancias['Teor Alcoólico (%)'], c=df_substancias['Cluster'], cmap='viridis')
plt.title('Agrupamento de Substâncias')
plt.xlabel('Concentração (%)')
plt.ylabel('Teor Alcoólico (%)')
plt.show()
```
