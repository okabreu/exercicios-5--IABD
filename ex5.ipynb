```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

dados = {
    'Máquina': ['M1', 'M2', 'M3', 'M4', 'M5', 'M6', 'M7', 'M8', 'M9', 'M10'],
    'Temperatura (°C)': [70.2, 65.1, 75.5, 80.3, 68.7, 72.9, 78.6, 66.4, 73.1, 69.5],
    'Vibração': [12.5, 8.2, 15.6, 10.2, 11.8, 14.3, 9.8, 8.9, 13.7, 12.1],
    'Corrente Elétrica': [4.7, 3.9, 5.1, 4.5, 4.2, 5.3, 4.8, 4.0, 5.0, 4.3]
}

df_maquinas = pd.DataFrame(dados)
print("DataFrame:")
print(df_maquinas)
```

```python
matriz_dados_maquinas = df_maquinas[['Temperatura (°C)', 'Vibração', 'Corrente Elétrica']].to_numpy()

kmeans = KMeans(n_clusters=2, random_state=42)
df_maquinas['Cluster'] = kmeans.fit_predict(matriz_dados_maquinas)

print("\nDataFrame com Clusters:")
print(df_maquinas)

sns.boxplot(x='Cluster', y='Temperatura (°C)', data=df_maquinas)
plt.title('Boxplot de Temperatura por Cluster')
plt.show()

sns.boxplot(x='Cluster', y='Vibração', data=df_maquinas)
plt.title('Boxplot de Vibração por Cluster')
plt.show()

sns.boxplot(x='Cluster', y='Corrente Elétrica', data=df_maquinas)
plt.title('Boxplot de Corrente Elétrica por Cluster')
plt.show()
```
